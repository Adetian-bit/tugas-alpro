```python
# Fungsi untuk menghitung penghasilan, pajak, dan pengeluaran Budi
def budi_finances(gaji_per_jam, jam_per_minggu):
    # Konstanta
    minggu = 5  # Total minggu Budi bekerja
    pajak = 0.14  # 14% pajak
    baju_aksesoris_persen = 0.10  # 10% dari pendapatan bersih
    alat_tulis_persen = 0.01  # 1% dari pendapatan bersih
    sedekah_persen = 0.25  # 25% dari sisa uang setelah belanja
    anak_yatim_persen = 0.30  # 30% dari sedekah untuk anak yatim
    dhuafa_persen = 0.70  # 70% dari sedekah untuk kaum dhuafa
    
    # 1. Total pendapatan sebelum pajak
    pendapatan_sebelum_pajak = gaji_per_jam * jam_per_minggu * minggu
    
    # 2. Pendapatan setelah pajak
    pendapatan_setelah_pajak = pendapatan_sebelum_pajak * (1 - pajak)
    
    # 3. Jumlah uang untuk membeli baju dan aksesoris
    pengeluaran_baju_aksesoris = pendapatan_setelah_pajak * baju_aksesoris_persen
    
    # 4. Jumlah uang untuk membeli alat tulis
    pengeluaran_alat_tulis = pendapatan_setelah_pajak * alat_tulis_persen
    
    # Uang yang tersisa setelah membeli baju, aksesoris, dan alat tulis
    sisa_uang_setelah_belanja = pendapatan_setelah_pajak - (pengeluaran_baju_aksesoris + pengeluaran_alat_tulis)
    
    # 5. Jumlah sedekah
    jumlah_sedekah = sisa_uang_setelah_belanja * sedekah_persen
    
    # 6. Jumlah yang diberikan kepada anak yatim
    uang_anak_yatim = jumlah_sedekah * anak_yatim_persen
    
    # 7. Jumlah yang diberikan kepada kaum dhuafa
    uang_dhuafa = jumlah_sedekah * dhuafa_persen
    
    # Outputkan hasil perhitungan
    hasil = {
        "Pendapatan sebelum pajak": pendapatan_sebelum_pajak,
        "Pendapatan setelah pajak": pendapatan_setelah_pajak,
        "Pengeluaran untuk baju dan aksesoris": pengeluaran_baju_aksesoris,
        "Pengeluaran untuk alat tulis": pengeluaran_alat_tulis,
        "Jumlah sedekah": jumlah_sedekah,
        "Uang untuk anak yatim": uang_anak_yatim,
        "Uang untuk kaum dhuafa": uang_dhuafa
    }
    
    return hasil

# Fungsi untuk menghitung diskon pembelian
def hitung_diskon(total_pembelian):
    diskon = 0
    if total_pembelian > 1000000:
        diskon = 0.1  # 10% diskon
    total_setelah_diskon = total_pembelian * (1 - diskon)
    
    return {
        "Diskon yang diterima": diskon * 100,
        "Total setelah diskon": total_setelah_diskon
    }

# Contoh input
gaji_per_jam = float(input("Masukkan gaji per jam: "))
jam_per_minggu = float(input("Masukkan jumlah jam kerja per minggu: "))
total_pembelian = float(input("Masukkan total pembelian: "))

# Hitung penghasilan Budi
hasil_budi = budi_finances(gaji_per_jam, jam_per_minggu)

# Hitung diskon pembelian
hasil_diskon = hitung_diskon(total_pembelian)

# Output hasil
print("\nHasil perhitungan Budi:")
for k, v in hasil_budi.items():
    print(f"{k}: {v}")

print("\nHasil diskon pembelian:")
for k, v in hasil_diskon.items():
    print(f"{k}: {v}")
```
